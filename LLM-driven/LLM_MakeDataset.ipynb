```python
from transformers import BlipProcessor, BlipForConditionalGeneration
import torch
from PIL import Image



try:
    processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
    model     = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

    def caption_pipeline(image: Image.Image) -> str:
        """
        Given a PIL image, returns a single string caption.
        """
        inputs  = processor(images=image, return_tensors="pt").to(device)
        out_ids = model.generate(**inputs, max_new_tokens=50)
        caption = processor.tokenizer.decode(out_ids[0], skip_special_tokens=True)
        return caption

    print("✅ BLIP caption pipeline ready!")
except Exception as e:
    caption_pipeline = None
    print("BLIP pipeline failed to load — relying on human captions.")
    print(e)


!pip install -U huggingface_hub sentence-transformers
from sentence_transformers import SentenceTransformer
embedder = SentenceTransformer(
    "multi-qa-MiniLM-L6-cos-v1",
    device=device,
    use_auth_token=True            # ← tells ST to re‑use the login() token
)
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base", use_fast=True)
llm = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base").to(device)
```

```python
#Version 2
# Colab cell ▽
def load_text(folder_path):
    """
    Load the article text from text.json.  Handles:
      - raw string
      - dict with a 'text' (or longest) string field
      - list of strings
    """
    path = os.path.join(folder_path, "text.json")
    data = json.load(open(path, encoding="utf-8"))

    # Case 1: it's already a raw string
    if isinstance(data, str):
        return data

    # Case 2: it's a dict
    if isinstance(data, dict):
        # try the obvious key first
        if "text" in data and isinstance(data["text"], str):
            return data["text"]
        # fallback: pick the longest string among the values
        candidates = [v for v in data.values() if isinstance(v, str)]
        if candidates:
            return max(candidates, key=len)
        # if you have a list somewhere in the dict, optionally handle it:
        for v in data.values():
            if isinstance(v, list):
                # flatten and join
                return "\n\n".join(str(x) for x in v)
        # last resort
        return str(data)

    # Case 3: it's a list of strings
    if isinstance(data, list):
        return "\n\n".join(str(x) for x in data)

    # Anything else: coerce to string
    return str(data)



import json, os

def load_captions(folder_path):
    """
    Read human‐written captions from img/meta.json under each topic folder.
    Handles the case where meta.json is itself a JSON string.
    """
    caps = {}
    meta_path = os.path.join(folder_path, "img", "meta.json")
    if not os.path.exists(meta_path):
        return caps

    # 1) Load the file
    raw = json.load(open(meta_path, encoding="utf-8"))

    # 2) If it's a string, parse it again
    if isinstance(raw, str):
        try:
            meta = json.loads(raw)
        except json.JSONDecodeError:
            print(f"Could not parse nested JSON in {meta_path}")
            return caps
    else:
        meta = raw

    # 3) meta should now be a dict with key "img_meta"
    for entry in meta.get("img_meta", []):
        if not isinstance(entry, dict):
            continue
        fname = entry.get("filename")
        desc  = entry.get("description", "").strip()
        if fname and desc:
            caps[fname] = desc

    return caps



def extract_with_llm(caption: str, paragraphs: list[str]) -> str:
    """
    Prompt Flan‐T5 to quote, verbatim, the single most on-topic fragment
    (up to 512 tokens) from these paragraphs that best describes the image.
    """
    prompt  = (
      "You are given an image caption and several paragraphs from an article.\n"
      "Identify which part of the *existing text* best describes the image, and *quote it verbatim*.\n"
      "Do not paraphrase or shorten it further—just return the exact excerpt (up to 512 tokens).\n\n"
      f"Image caption: {caption}\n\n"
      "Paragraphs:\n"
    )
    for i,p in enumerate(paragraphs, 1):
        prompt += f"{i}. {p}\n\n"
    prompt += "Answer with the excerpt exactly as it appears."

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=1024
    ).to(device)

    out = llm.generate(
        **inputs,
        max_new_tokens=512,
        do_sample=False
    )
    return tokenizer.decode(out[0], skip_special_tokens=True).strip()

def chunk_windows(text: str, window: int = 200, stride: int = 100) -> list[str]:
    """
    Break `text` into overlapping windows of ~`window` words, stepping by `stride`.
    """
    words = text.split()
    chunks = []
    for i in range(0, max(1, len(words) - window + 1), stride):
        chunks.append(" ".join(words[i : i + window]))
    # include the last window if anything remains
    if len(words) > window and (i + window) < len(words):
        chunks.append(" ".join(words[-window:]))
    return chunks or [" ".join(words)]

```

```python
# ── Before this cell, make sure you have:
#   - defined load_text(...)
#   - defined load_captions(...)
#   - defined chunk_windows(...)
#   - defined extract_with_llm(...) with your "QUOTE IT VERBATIM" prompt
#   - defined caption_pipeline(...) and loaded device, tokenizer, llm, etc.

results = []

for topic in folders:
    folder    = os.path.join(base_path, topic)
    full_text = load_text(folder)                   # the entire wikipedia passage
    windows   = chunk_windows(full_text, window=200, stride=100)
    captions  = load_captions(folder)

    for img in os.listdir(os.path.join(folder, "img")):
        if not img.lower().endswith((".jpg",".png")):
            continue
        img_path = os.path.join(folder, "img", img)

        # 1) Get or generate a caption
        cap = captions.get(img, "")
        if not cap and caption_pipeline:
            cap = caption_pipeline(
                Image.open(img_path).convert("RGB")
            )
        if not cap:
            print(f"⚠️  Skipping {img}: no caption found.")
            continue
        cap = cap.replace("\n", " ").strip()

        # 2) LLM distillation on full-text windows
        #    We give the LLM all 200-word slices at once.
        snippet = extract_with_llm(cap, windows)

        results.append({
            "image": os.path.join(topic, "img", img),
            "text_segment": snippet
        })

# Summary
print(f"Extracted {len(results)} pairs.")
```

```python
import pandas as pd

df = pd.DataFrame(results)
df.to_csv("image_text_pairs.csv", index=False)
print(f"✅ Saved {len(df)} rows to image_text_pairs.csv")
```
